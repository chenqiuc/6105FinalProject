```python
import datetime
def dataPrepare(date, d):
    date = datetime.datetime.strptime(date, '%Y-%m-%d')
    delta = datetime.timedelta(days=d)
    n_days = date + delta
    newDate = n_days.strftime('%Y-%m-%d')
    return newDate
```
